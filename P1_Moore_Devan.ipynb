##TASK 3
###Column by Column Exploration

```python
y_target = df.pop('y')
```

```python
for col in df.columns:
  print(f"\nAnalysis of Column: {col}")
  print(f"Data Type: {df[col].dtype}")

  if pd.api.types.is_numeric_dtype(df[col]):
    print(f"Mean: {df[col].mean().round(4)}")
    print(f"Median: {df[col].median()}")
    print(f"Standard Deviation: {df[col].std()}")
    print(f"Minimum Value: {df[col].min()}")
    print(f"Maximum Value: {df[col].max()}")
    print(f"Number of Unique Values: {df[col].nunique()}")
    print(f"Value Counts: \n{df[col].value_counts()}")

    sns.histplot(df[col])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
    print("\n")

  elif pd.api.types.is_object_dtype(df[col]) or pd.api.type.is_categorical_dtype(df[col]):
    print(f"Value Counts: {df[col].value_counts()}")

    sns.countplot(df[col])
    plt.title(f'Counts of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
    print("\n")

  else:
    print(f"Unable to process Data for: {col}")
```

```python
print(f"Original DataFrame Shape: {df.shape}")
print(f"Encoded DataFrame Shape: {df_encoded.shape}")
```

One-hot encoding is a way to turn categorical variables in the data into numbers that machine learning models can understand. For this analysis, I applied one-hot encoding to several categorical columns in our DataFrame, including 'workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', and 'sex'.

The DataFrame started off with (32561, 14), meaning it had 32,561 rows and 14 columns. However, after I applied one-hot encoding, the shape changed to (32561, 108). This increase in columns happened because each categorical variable gets turned into multiple columns, with each new column showing whether a category is true or not.
For example, the marital-status column could turn into new columns like marital-status_Married and marital-status_Single, which makes it easier for models to work with.

In conclusion, one-hot encoding helps models handle categorical data better by converting it into a numerical format. This process adds more columns to the DataFrame but also makes the data more useful for analysis and predictions.
